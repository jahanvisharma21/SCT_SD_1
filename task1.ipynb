```python
print("Temperature Converter")
print("Enter the temperature scale you want to convert FROM:")
print("C - Celsius")
print("F - Fahrenheit")
print("K - Kelvin")

from_scale = input("Enter scale (C/F/K): ").upper()
value = float(input("Enter temperature value: "))

print("\nEnter the temperature scale you want to convert TO:")
print("C - Celsius")
print("F - Fahrenheit")
print("K - Kelvin")

to_scale = input("Enter scale (C/F/K): ").upper()

# Dictionary to print the full form
scale_names = {
    "C": "Celsius",
    "F": "Fahrenheit",
    "K": "Kelvin"
}

def convert_temperature(value, from_scale, to_scale):
    # convert input to Celsius first
    if from_scale == "C":
        celsius = value
    elif from_scale == "F":
        celsius = (value - 32) * 5/9
    elif from_scale == "K":
        celsius = value - 273.15
    else:
        return None

    # convert Celsius to output
    if to_scale == "C":
        return celsius
    elif to_scale == "F":
        return (celsius * 9/5) + 32
    elif to_scale == "K":
        return celsius + 273.15
    else:
        return None


result = convert_temperature(value, from_scale, to_scale)

if result is None:
    print("Invalid scale entered!")
else:
    print("\n====== RESULT ======")
    print(f"Selected From Scale : {scale_names.get(from_scale, 'Invalid')}")
    print(f"Selected To Scale   : {scale_names.get(to_scale, 'Invalid')}")
    print(f"Input Temperature   : {value} {scale_names[from_scale]}")
    print(f"Converted Value     : {result:.2f} {scale_names[to_scale]}")

```
